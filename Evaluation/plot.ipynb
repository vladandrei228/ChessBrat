```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file with blunder rates
csv_file_path = './br/BR1800.csv'
df = pd.read_csv(csv_file_path)

# Ensure the columns are in the correct order and convert data to numeric types
columns = ['Format', 'Actual', 'ChessBrat', 'MaiaChess', 'Leela', 'Stockfish']
df = df[columns]
df[columns[1:]] = df[columns[1:]].apply(lambda x: pd.to_numeric(x.str.replace('%', ''), errors='coerce'))

# Replace NaN/Inf values with 0
df.fillna(0, inplace=True)
df.replace([float('inf'), -float('inf')], 0, inplace=True)

# Extract the formats and blunder rates
formats = df['Format']
actual_blunder_rate = df['Actual']
chessbrat_blunder_rate = df['ChessBrat']
maiachess_blunder_rate = df['MaiaChess']
leela_blunder_rate = df['Leela']
stockfish_blunder_rate = df['Stockfish']

# Create a grouped bar plot
fig, ax = plt.subplots(figsize=(14, 8))
width = 0.15  # Width of the bars
x = np.arange(len(formats))

# Plot bars for each engine and actual moves
ax.bar(x - 2 * width, actual_blunder_rate, width, label='Actual', color='purple')
ax.bar(x - width, chessbrat_blunder_rate, width, label='ChessBrat', color='red')
ax.bar(x, maiachess_blunder_rate, width, label='MaiaChess', color='orange')
ax.bar(x + width, leela_blunder_rate, width, label='Leela', color='green')
ax.bar(x + 2 * width, stockfish_blunder_rate, width, label='Stockfish', color='blue')

# Add labels and title
ax.set_ylabel('Blunder Rate (%)')
ax.set_title('Blunder Rates (1800 ELO Rating)')
ax.set_xticks(x)
ax.set_xticklabels(formats)
ax.set_xlabel('Formats')

# Add grid lines
ax.yaxis.grid(True, linestyle='--', linewidth=0.7)

# Add legend
ax.legend()

# Add the values on top of the bars
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.2f%%', label_type='edge', padding=3)

# Adjust y-axis limits dynamically
max_blunder_rate = max(
    np.nanmax(actual_blunder_rate), 
    np.nanmax(chessbrat_blunder_rate), 
    np.nanmax(maiachess_blunder_rate), 
    np.nanmax(leela_blunder_rate), 
    np.nanmax(stockfish_blunder_rate)
)
ax.set_ylim(0, max_blunder_rate + 1)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file with MMR
csv_file_path = './mmr/mmr1800.csv'
df = pd.read_csv(csv_file_path)

# Ensure the columns are in the correct order and convert data to numeric types
columns = ['Format', 'ChessBrat', 'MaiaChess', 'Leela', 'Stockfish']
df = df[columns]

# Convert the data to numeric types
for col in columns[1:]:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace('%', ''), errors='coerce')

# Replace NaN/Inf values with 0
df.fillna(0, inplace=True)
df.replace([float('inf'), -float('inf')], 0, inplace=True)

# Extract the formats and MMR values
formats = df['Format']
chessbrat_mmr = df['ChessBrat']
maiachess_mmr = df['MaiaChess']
leela_mmr = df['Leela']
stockfish_mmr = df['Stockfish']

# Create a grouped bar plot
fig, ax = plt.subplots(figsize=(14, 8))
width = 0.15  # Width of the bars
x = np.arange(len(formats))

# Plot bars for each engine
bars1 = ax.bar(x - 1.5 * width, chessbrat_mmr, width, label='ChessBrat', color='red')
bars2 = ax.bar(x - 0.5 * width, maiachess_mmr, width, label='MaiaChess', color='orange')
bars3 = ax.bar(x + 0.5 * width, leela_mmr, width, label='Leela', color='green')
bars4 = ax.bar(x + 1.5 * width, stockfish_mmr, width, label='Stockfish', color='blue')

# Add labels and title
ax.set_ylabel('MMR (%)')
ax.set_title('Move Matching Rate (1800 ELO Rating)')
ax.set_xticks(x)
ax.set_xticklabels(formats)
ax.set_xlabel('Formats')

# Add grid lines
ax.yaxis.grid(True, linestyle='--', linewidth=0.7)

# Add legend
ax.legend()

# Add the values on top of the bars
ax.bar_label(bars1, fmt='%.2f%%', padding=3)
ax.bar_label(bars2, fmt='%.2f%%', padding=3)
ax.bar_label(bars3, fmt='%.2f%%', padding=3)
ax.bar_label(bars4, fmt='%.2f%%', padding=3)

# Adjust y-axis limits dynamically
max_mmr = max(
    np.nanmax(chessbrat_mmr), 
    np.nanmax(maiachess_mmr), 
    np.nanmax(leela_mmr), 
    np.nanmax(stockfish_mmr)
)
ax.set_ylim(0, max_mmr + 1)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('TER1800.csv')

# Define colors for each format
colors = {
    'Bullet': 'red',
    'Blitz': 'orange',
    'Rapid': 'green',
    'Classical': 'blue',
    'Overall': 'purple'
}

# Create a soil plot with bars
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(df['Format'], df['TER'], color=[colors[fmt] for fmt in df['Format']], edgecolor='black')

# Add dot plot on top of bars and write numbers
for bar in bars:
    height = bar.get_height()
    ax.plot(bar.get_x() + bar.get_width() / 2, height, 'o', color='black')
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

# Add labels and title
ax.set_xlabel('Format')
ax.set_ylabel('Time Efficiency Ratio (TER)')
ax.set_title('Time Efficiency Ratio (TER) at 1800 ELO')

# Remove the grid
ax.grid(False)

# Show the plot
plt.tight_layout()
plt.show()
```
